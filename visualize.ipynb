```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = np.loadtxt('data.csv', delimiter=',')
filters, n = data.shape
time = np.arange(n)

plt.figure(figsize=(10, 6))
for i in range(filters):
    plt.plot(time, data[i, :], label=f'Filter {i + 1}')

plt.xlabel('Time')
plt.ylabel('Measurement Value')
plt.title('Visualization of Filter Outputs')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Estimate
estimate = np.loadtxt('estimate.csv', delimiter=',')
filters, n = estimate.shape
time = np.arange(n)

plt.figure(figsize=(10, 6))
plt.plot(time, data[0, :], label='Measurement')
plt.plot(time, estimate[0, :], c='r', label='Estimate')

plt.xlabel('Time')
plt.ylabel('Estimate Measurement Value')
plt.title('Visualization of UKF Outputs (Filter 1)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
